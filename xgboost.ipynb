```python
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization

def build_model2(hp):
    model2 = Sequential()

    num_layers = hp.Int('num_layers', min_value=1, max_value=10)

    for i in range(num_layers):
        # First hidden layer
        if i == 0:
            model2.add(
                Dense(
                    hp.Int('units' + str(i), min_value=8, max_value=128, step=8),
                    activation=hp.Choice('activation' + str(i), values=['relu', 'tanh']),
                    input_dim=4
                )
            )
            model2.add(Dropout(0.3),BatchNormalization())
        # Other hidden layers
        else:
            model2.add(
                Dense(
                    hp.Int('units' + str(i), min_value=8, max_value=128, step=8),
                    activation=hp.Choice('activation' + str(i), values=['relu', 'tanh'])
                )
            )
            model2.add(Dropout(0.2))
            

    # Output layer (for classification)
    model2.add(Dense(1, activation='sigmoid'))

    # Compile model
    model2.compile(
        optimizer=hp.Choice('optimizer', values=['Adam', 'rmsprop', 'SGD']),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model2
```

```python
tuner=kt.RandomSearch(build_model2,objective='accuracy',max_trials=20,directory='my_dir',
    project_name='my_project6.5'
)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam, RMSprop, SGD
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import keras_tuner as kt

def build_model2(hp):
    model2 = Sequential()

    # Tune number of hidden layers
    num_layers = hp.Int('num_layers', min_value=1, max_value=6)

    for i in range(num_layers):
        # Number of units per layer
        units = hp.Int('units_' + str(i), min_value=16, max_value=256, step=16)

        # Add dense layer
        if i == 0:
            model2.add(Dense(units, activation='relu', input_dim=10))
        else:
            model2.add(Dense(units, activation='relu'))

        # BatchNorm and Dropout
        model2.add(BatchNormalization())
        model2.add(Dropout(hp.Float('dropout_' + str(i), min_value=0.1, max_value=0.4, step=0.1)))

    # Output layer (Binary classification)
    model2.add(Dense(1, activation='sigmoid'))

    # Tune optimizer and learning rate
    optimizer_choice = hp.Choice('optimizer', ['adam', 'rmsprop', 'sgd'])
    learning_rate = hp.Float('learning_rate', min_value=1e-4, max_value=1e-2, sampling='log')

    if optimizer_choice == 'adam':
        optimizer = Adam(learning_rate=learning_rate)
    elif optimizer_choice == 'rmsprop':
        optimizer = RMSprop(learning_rate=learning_rate)
    else:
        optimizer = SGD(learning_rate=learning_rate, momentum=0.9)

    # Compile model
    model2.compile(
        optimizer=optimizer,
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model2

# --- Tuner ---
tuner = kt.RandomSearch(
    build_model2,
    objective='val_accuracy',
    max_trials=25,
    directory='my_dir',
    project_name='my_project5056'
)

# --- Callbacks ---
callbacks = [
    EarlyStopping(monitor='val_accuracy', patience=25, restore_best_weights=True, min_delta=0.005, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10, min_lr=1e-6, verbose=1)
]

# --- Search best model ---
tuner.search(
    x_train_scaled1, y_train1,
    validation_data=(x_test_scaled1, y_test1),
    epochs=300,
    callbacks=callbacks,
    verbose=1
)
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model3= tuner.get_best_models(num_models=1)[0]
```

```python
model3.summary()
```

```python
model3.fit(x_train_scaled1,y_train1,epochs=500,initial_epoch=301,validation_data=(x_test_scaled1,y_test1),verbose=True)
```

```python
y_log3=model3.predict(x_test_scaled1)
```
